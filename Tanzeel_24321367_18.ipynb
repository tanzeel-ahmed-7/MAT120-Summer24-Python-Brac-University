```python
#Question no.1 solve

import sympy as sp

#variable of function f(t)
t = sp.symbols('t')
f = 50*t**3 - 36*t**2 - 63*t + 18

#derivative of f with respect to t
df_dt = sp.diff(f, t)

#evaluation of the derivative
df_dt_at_0 = df_dt.subs(t, 0)

print(f"df/dt at t=0: {df_dt_at_0}")
```

```python
#Question no.2 solve

import numpy as np
import matplotlib.pyplot as plt

#Using Euler's method

def euler_method(f, g, y0, v0, t0, tf, h):

#Specific time points
  t= np.arange(t0, tf+h, h)

#Solution store arrays
  y= np.zeros(len(t))
  v= np.zeros(len(t))

#Begining conditions
  y[0]= y0
  v[0]= v0

#Loop of Euler's method
  for i in range(1, len(t)):
    v[i] = v[i-1]+h*g(y[i-1], v[i-1], t[i-1])
    y[i] = y[i-1]+h*f(y[i-1], v[i-1], t[i-1])

  return t, y

#Function for dy/dt=v
def f(y, v, t):
  return v

#Function for dv/dt=6v-25y
def g(y, v, t):
  return 6*v-25*y

#Begining conditions

y0 = 0  # Initial position
v0 = 1  # Initial velocity (for example)
t0 = 0  # Initial time
tf = 4  # Final time
h = 0.01 # Step size

#Solving the problem using Euler's Method

t, y = euler_method(f, g, y0, v0, t0, tf, h)

# Result plot
plt.plot(t, y, label="y(t)")
plt.xlabel("Time(t)")
plt.ylabel("y(t)")
plt.title("Particular Solution for Euler's Method")
plt.grid(True)
plt.legend()
plt.show()
```
